### Odds de cada partida do campeonato

### Vamos juntar os dois data frames, matches e odds, para podermos saber como estao as espectativas pre jogo

```python
# primeiro precisamos unificar os nomes dos times que estao diferentes em ambos os data frames

# Renomear as colunas em odds_df para corresponder às colunas em data
odds_df = odds_df.rename(columns={
    'HomeTeam': 'home_team_name',
    'AwayTeam': 'away_team_name',
})

# Criar um dicionário de mapeamento para padronizar os nomes dos times
team_name_mapping = {
    'Man United': 'Manchester United',
    'Man City': 'Manchester City',
    'Leicester': 'Leicester City',
    'Newcastle': 'Newcastle United',
    'Bournemouth': 'AFC Bournemouth',
    'Brighton': 'Brighton & Hove Albion',
    'Wolves': 'Wolverhampton Wanderers',
    'West Ham': 'West Ham United',
    'Cardiff': 'Cardiff City',
    'Tottenham': 'Tottenham Hotspur',
    'Huddersfield': 'Huddersfield Town',
}


# Aplicar o mapeamento aos DataFrames
matches_df['home_team_name'] = matches_df['home_team_name'].replace(team_name_mapping)
matches_df['away_team_name'] = matches_df['away_team_name'].replace(team_name_mapping)
odds_df['home_team_name'] = odds_df['home_team_name'].replace(team_name_mapping)
odds_df['away_team_name'] = odds_df['away_team_name'].replace(team_name_mapping)

# Realizar a junção dos DataFrames com base nos nomes padronizados dos times
matches_info_df = pd.merge(matches_df, odds_df, on=['home_team_name', 'away_team_name'], how='inner')
matches_info_df
```

### Criação das funções que iremos utilizar para trabalhar melhor os dados

```python
def checking_result(row):
    result = row['home_team_goal_count'] - row['away_team_goal_count']
        
    return result

# defining the result
matches_info_df['result'] = matches_info_df.apply(checking_result, axis=1)

# Divide the season into three parts based on index position
matches_info_df["season_part"] = 'first'
matches_info_df.loc[(matches_info_df.index > 120) & (matches_info_df.index < 240),'season_part'] = 'mid'
matches_info_df.loc[(matches_info_df.index >= 240),'season_part'] = 'last'

# Converting the 'match_datetime' column to datetime format
matches_info_df['match_datetime'] = pd.to_datetime(matches_info_df['date_GMT'], format='%b %d %Y - %I:%M%p')
# Creating the 'day_or_night' column based on the hour of the match
matches_info_df['day_or_night'] = matches_info_df['match_datetime'].dt.hour.apply(lambda x: 'Night' if x >= 17 else 'Day')

def home_second_half_goals(row):
    home_goals_second_half = row['home_team_goal_count'] - row['home_team_goal_count_half_time']
    
    return int(home_goals_second_half)

def away_second_half_goals(row):
    away_goals_second_half = row['away_team_goal_count'] - row['away_team_goal_count_half_time']
    
    return int(away_goals_second_half)

# creating column to goals on 2 half
matches_info_df['home_team_goal_count_secondhalf_time'] = matches_info_df.apply(home_second_half_goals, axis=1)
matches_info_df['away_team_goal_count_secondhalf_time'] = matches_info_df.apply(away_second_half_goals, axis=1)

# Function to verify if there are goals scored after the 80th minute for both teams
def has_goals_after_80_minutes(row):
    home_goals = row['home-80-final']
    away_goals = row['away-80-final']
    
    return home_goals or away_goals

# Function to convert time notation to minutes
def convert_time_notation_to_minutes(time_str):
    values = time_str.split("'")
    if len(values) != 1 and len(values) != 2:
        raise Exception('Invalid time notation')

    minutes = int(values[0])
    if len(values) == 2:
        minutes += int(values[1])

    return minutes

# Function to check if there are goals scored after a specified threshold time
def has_goals_after_threshold(times_str, threshold=80):
    # Splitting the goal timings string using comma delimiter
    timings_arr = times_str.split(',')
    # Converting time notation to minutes
    minutes_arr = [convert_time_notation_to_minutes(x) for x in timings_arr]

    # Checking if any time is greater than or equal to the threshold
    for time in minutes_arr:
        if time >= threshold:
            return True
    
    return False

# Applying the has_goals_after_threshold function to home and away goal timings, and combining the results
matches_info_df['home-80-final'] = matches_info_df['home_team_goal_timings'].apply(lambda x: has_goals_after_threshold(x) if not isinstance(x, float) else False)
matches_info_df['away-80-final'] = matches_info_df['away_team_goal_timings'].apply(lambda x: has_goals_after_threshold(x) if not isinstance(x, float) else False)
matches_info_df['80-final'] = matches_info_df.apply(has_goals_after_80_minutes, axis=1)

# Selecting relevant columns for analysis
matches_info_df[['home_team_goal_timings', 'away_team_goal_timings', 'home-80-final', 'away-80-final', '80-final', 'home_team_goal_count_secondhalf_time']].head(20)
```

```python
# Adicionar coluna de índice original ao DataFrame matches_info_df
matches_info_df['original_index'] = matches_info_df.index

# Realizar a junção dos DataFrames com base nos nomes padronizados dos times
teams_stats.rename(columns={'common_name': 'home_team_name', 'shots': 'home_season_shots', 'shots_on_target': 'home_season_shots_on_target', 'performance_rank': 'performance_rank_home'}, inplace=True)
matches_info_df = pd.merge(matches_info_df, teams_stats, on=['home_team_name'], how='inner')

teams_stats = teams_stats.drop(columns=['home_season_shots', 'home_season_shots_on_target'])
teams_stats.rename(columns={'home_team_name': 'away_team_name', 'performance_rank_home': 'away_performance_rank'}, inplace=True)
matches_info_df = pd.merge(matches_info_df, teams_stats, on=['away_team_name'], how='inner')

# Ordenar pelo índice original para manter a ordem das linhas
matches_info_df = matches_info_df.sort_values(by='original_index').reset_index(drop=True)

# Remover a coluna de índice original
matches_info_df = matches_info_df.drop(columns=['original_index'])

matches_info_df

# falar com o professor para ver se isso aqui é merda
```

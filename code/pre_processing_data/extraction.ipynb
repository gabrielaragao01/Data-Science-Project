## Deixando apenas os campos que vamos usar na analise (dropando o que nao vamos usar)

```python
# definindo colunas que vamos usar:
matches_columns = ['date_GMT', 'home_team_name', 'away_team_name', 'Pre-Match PPG (Home)', 'Pre-Match PPG (Away)', 'home_team_goal_count', 'away_team_goal_count', 'home_team_goal_count_half_time', 'away_team_goal_count_half_time', 'home_team_goal_timings', 'away_team_goal_timings', 'home_team_shots', 'home_team_shots_on_target', 'away_team_shots', 'away_team_shots_on_target']

matches_df = matches_df[matches_columns]
matches_df
```

```python
odds_columns = ['HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A']

odds_df = odds_df[odds_columns]
odds_df
```

## Vamos juntar os dois data frames, matches e odds, para podermos saber como estao as espectativas pre jogo

```python
# primeiro precisamos unificar os nomes dos times
#que estao diferentes em ambos os data frames

# Renomear as colunas em odds_df para corresponder às colunas em data
odds_df = odds_df.rename(columns={
    'HomeTeam': 'home_team_name',
    'AwayTeam': 'away_team_name'
})

# Criar um dicionário de mapeamento para padronizar os nomes dos times
team_name_mapping = {
    'Man United': 'Manchester United',
    'Man City': 'Manchester City',
    'Leicester': 'Leicester City',
    'Newcastle': 'Newcastle United',
    'Bournemouth': 'AFC Bournemouth',
    'Brighton': 'Brighton & Hove Albion',
    'Wolves': 'Wolverhampton Wanderers',
    'West Ham': 'West Ham United',
    'Cardiff': 'Cardiff City',
    'Tottenham': 'Tottenham Hotspur',
    'Huddersfield': 'Huddersfield Town',
}


# Aplicar o mapeamento aos DataFrames
matches_df['home_team_name'] = matches_df['home_team_name'].replace(team_name_mapping)
matches_df['away_team_name'] = matches_df['away_team_name'].replace(team_name_mapping)
odds_df['home_team_name'] = odds_df['home_team_name'].replace(team_name_mapping)
odds_df['away_team_name'] = odds_df['away_team_name'].replace(team_name_mapping)

# Realizar a junção dos DataFrames com base nos nomes padronizados dos times
matches_info_df = pd.merge(matches_df, odds_df, on=['home_team_name', 'away_team_name'], how='inner')
matches_info_df
```

```python
# criando mais duas colunas ## PROFESSOR OLHAR PQ TO PEGANDO POR INDICE NAO DATA
matches_info_df["season_part"] = 'first'
matches_info_df.loc[(matches_info_df.index > 120) & (matches_info_df.index < 240),'season_part'] = 'mid'
matches_info_df.loc[(matches_info_df.index >= 240),'season_part'] = 'last'

# Converter a coluna 'match_datetime' para datetime
matches_info_df['match_datetime'] = pd.to_datetime(matches_info_df['date_GMT'], format='%b %d %Y - %I:%M%p')
# Criar a coluna 'time_of_day'
matches_info_df['day_or_night'] = matches_info_df['match_datetime'].dt.hour.apply(lambda x: 'Night' if x >= 17 else 'Day')

def time_notation_to_time(time_str):
    values = time_str.split("'")
    if len(values) != 1 and len(values) != 2:
        raise Exception('Bad time notation')

    value = int(values[0])
    if len(values) == 2:
        value += int(values[1])

    return value

def has_goals_after(times_str, threshold = 80):
    # Dividir a string do gol usando split
    times_str_arr = times_str.split(',')
    #regex
    times_arr = [time_notation_to_time(x) for x in times_str_arr]

    # Verificar se algum tempo é maior que 80
    for time in times_arr:
        if time >= threshold:
            return True
    
    return False

matches_info_df['80-final'] = matches_info_df['away_team_goal_timings'].apply(lambda x: has_goals_after(x) if type(x) != float else 0)

'''
# criando tabela de gols nos ultimos 10 min
matches_info_df["goal_80-final"] = 'no'
matches_info_df.loc[(int(matches_info_df['home_team_goal_timings']) >= 80) or (int(matches_info_df['away_team_goal_timings']) >= 80), 'goal_80-final'] = 'yes'
'''
matches_info_df
```

```python
## colocando as colunas na ordem correta
ordered_df = ['B365H','home_team_name','B365D', 'away_team_name', 'B365A', 'Pre-Match PPG (Home)', 'Pre-Match PPG (Away)', 'home_team_goal_count', 'away_team_goal_count', 'home_team_goal_count_half_time', 'away_team_goal_count_half_time', 'home_team_goal_timings', 'away_team_goal_timings', 'home_team_shots', 'home_team_shots_on_target', 'away_team_shots', 'away_team_shots_on_target', 'season_part', 'day_or_night']

matches_info_df = matches_info_df[ordered_df]
matches_info_df.to_csv('matches_info_df.csv')
matches_info_df
```

```python
import pandas as pd
import numpy as np
```

```python
matches_info_df = pd.read_csv("matches_info_df.csv", index_col=0)
matches_info_df
```

```python
porcentagem = 0.5
num_linhas = len(matches_info_df)
num_valores_para_sujar = int(num_linhas * porcentagem)

coluna = 'home_team_shots_on_target'

indices_para_sujar = np.random.choice(matches_info_df.index, num_valores_para_sujar, replace=False)

matches_info_df.loc[indices_para_sujar, coluna] = np.nan

num_nans = matches_info_df['home_team_shots_on_target'].isna().sum()

print(num_nans)
matches_info_df.tail(50)
```
